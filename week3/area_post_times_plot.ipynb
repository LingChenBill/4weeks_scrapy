整理每个地区出现的频率，发帖数

```python
area_list = []
for i in item_info.find():
    if len(i['area']) > 1:
        area_list.append(i['area'][1])
area_index = list(set(area_list))
print(area_index)
```

整理每个地区的发帖数

```python
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(post_times)
```

结构生成器:series

```python
def data_gen_series(types):
    length = 0
    if length <= len(area_index):
        for area, times in zip(area_index, post_times):
            data = {
                'name': area,
                'data': [times],
                'type': types
            }
            yield data
            length += 1
```

```python
data_gen_series('column')
```

```python
for i in data_gen_series('column'):
    print(i)
```

```python
series = [data for data in data_gen_series('column')]
charts.plot(series, show='inline', options=dict(title=dict(text='七日内北京城区二手物品发帖量')))
```
