```python
import pymongo
import charts
```

```python
client = pymongo.MongoClient('localhost', 27017)
ganji = client['ganji']
url_list = ganji['url_list']
item_info = ganji['item_infoY']
```

```python
for i in url_list.find().limit(300):
    print(i['url'])
```

观察url链接特征,提取二手类目到列表中

```python
category_url=[]
for i in url_list.find():
    category_url.append(i['url'].split('/')[-2])
category_index = list(set(category_url))
print(category_index)
```

名称转换

```python
category_name = {
    'jiaju': '二手家具',
    'rirongbaihuo': '家居百货',
    'jiadian': '家用电器',
    'meironghuazhuang': '美容护肤/化妆品',
    'shuma': '电子数码',
    'bangong': '设备/办公用品',
    'shouji': '二手手机',
    'xianzhilipin': '闲置礼品',
    'diannao': '台式电脑/网络',
    'ruanjiantushu': '图书/音乐/运动',
    'shoucangpin': '收藏品',
    'baojianpin': '保健品',
    'ershoubijibendiannao': '二手平板电脑'
}
jiaju = 'jiaju'
print(category_name[jiaju])
```

二手类目的发帖数

```python
category_times = []
for i in category_index:
    category_times.append(category_url.count(i))
print(category_times)
```

图片生成器函数

```python
def category_gen_series(types):
    length = 0
    if length <= len(category_index):
        for category, times in zip(category_index, category_times):
            data = {
                'name': category_name[category],
                'data': [times],
                'type': types
            }
            yield data
            length += 1
```

```python
for i in category_gen_series('column'):
    print(i)
```

绘制图表

```python
series = [data for data in category_gen_series('column')]
charts.plot(series, show='inline', options=dict(title=dict(text='二手类目的发帖数')))
```
