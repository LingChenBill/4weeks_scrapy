添加times字段，每条数据的发贴使用天数

```python
for i in item_info.find():
    print(i['_id'])
    time = random.randint(1, 10)
    item_info.update({'_id': i['_id']}, {'$set': {'times': time}})
```

```python
for i in item_info.find().limit(200):
    print(i)
```

# 使用管道

##  {'$match': {'times': 33}}

## {'$match': {'$and': [{'publish_date': '2019.03.28'}, {'times': 33}]}}

```python
pipeline = [
    {'$match': {'times': 5}},
    {'$group': {'_id': '$price', 'counts': {'$sum': 1}}},
    {'$sort': {'counts': -1}},
    {'$limit': 2}
]
```

```python
for i in item_info.aggregate(pipeline):
    print(i)
```
